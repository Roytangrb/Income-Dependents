```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook

df = pd.read_csv('农村居民中等偏上户平均每个劳动力负担人口.csv')
'''
df.columns: ['种类列', 'region', 'unit', 'time', 'regionName', 'value', 'type']
df.index.unique(): ['农村居民中等偏上户平均每个劳动力负担人口', '农村居民低收入户平均每个劳动力负担人口', '农村居民中等收入户平均每个劳动力负担人口',
       '农村居民高收入户平均每个劳动力负担人口', '农村居民平均每个劳动力负担人口(含本人)', '农村居民中等偏下户平均每个劳动力负担人口']
'''
df = df.reset_index()
df.columns = ['种类列', 'region', 'unit', 'time', 'regionName', 'value', 'type', 'dummy']
df = df[['种类列', 'unit', 'time', 'regionName', 'value']]
df
```

```python
import numpy as np

df_high = df[df['种类列'] == '农村居民高收入户平均每个劳动力负担人口']
df_low = df[df['种类列'] == '农村居民低收入户平均每个劳动力负担人口']
df_upper = df[df['种类列'] == '农村居民中等偏上户平均每个劳动力负担人口']
df_lower = df[df['种类列'] == '农村居民中等偏下户平均每个劳动力负担人口']
df_middle = df[df['种类列'] == '农村居民中等收入户平均每个劳动力负担人口']
df_avg = df[df['种类列'] == '农村居民中等偏上户平均每个劳动力负担人口']

plt.plot(df_avg['time'], df_avg['value'], '--', c='lightgreen', label='Average')
plt.plot(df_high['time'], df_high['value'], '-o', c='black', alpha=0.6, label='High Income')
plt.plot(df_upper['time'], df_upper['value'], '-o', c='blue',alpha=0.7, label='Medium Upper Income')
plt.plot(df_middle['time'], df_middle['value'], '-o', c='blue',alpha=0.5, label='Medium Income')
plt.plot(df_lower['time'], df_lower['value'], '-o', c='blue',alpha=0.3, label='Medium Lower Income')
plt.plot(df_low['time'], df_low['value'], '-o', c='blue',alpha=0.1, label='Low Income')

plt.title('Number of dependents per labour force in rural areas, China')
plt.xlabel('Year')
plt.ylabel('Number of dependents (self included)')
plt.legend(loc=1, frameon=False)

ax = plt.gca()
_ = ax.axis([2001, 2013, 1, 2])
_ = plt.xticks(np.arange(2002, 2013, 1))


```
